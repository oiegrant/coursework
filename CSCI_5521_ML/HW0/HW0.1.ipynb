```python
import scipy.io
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
mat = scipy.io.loadmat("data1.mat")
```

```python
X, y = mat['X'], mat['y']

X = np.array(X)
# plt.scatter(X[:,0],X[:,1],c=y)
mxx = max(X[:,0])
mnx = min(X[:,0])
mxy = max(X[:,1])
mny = min(X[:,1])
w = [1,-1]
x1 = np.arange(mnx,mxx,0.01)
x2 = -(w[0]*x1 + w[1])/w[1]
plt.clf()
plt.scatter(X[:, 0], X[:, 1],c=y)
plt.plot(x1,x2,'y-')
plt.xlim([mnx,mxx])
plt.ylim([mny,mxy])
```

```python
w_init = [1,-1]
mxx = max(X[:,0])
mnx = min(X[:,0])
mxy = max(X[:,1])
mny = min(X[:,1])


def MyPerceptron(data, labels, w0):
    err = 1
    round = 0
    w = w0

    while err > 0:
        for i in range(len(data)):
            if (np.dot(data[i],w)*labels[i]) <= 0:
                w = w + (labels[i] * data[i])
                x1 = np.arange(mnx,mxx,0.01)
                x2 = -(w[0]*x1 + w[1])/w[1]
                plt.clf()
                plt.scatter(data[:, 0], data[:, 1],c=labels)
                plt.plot(x1,x2,'y-')
                plt.xlim([mnx,mxx])
                plt.ylim([mny,mxy])


        round += 1
        err = np.sum(np.sign(np.dot(data,w.T)) != labels) / len(labels)

    return w, round
                
MyPerceptron(X,y,w_init)    
```
